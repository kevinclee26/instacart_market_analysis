# Instacart Market Analysis

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as sts

subplot_cols=2
chart_h=5
max_bars=200
aspect_ratio=4/3
```

```python
files=os.listdir('instacart_2017_05_01')
[print(each_file) for each_file in files]
dfs=[pd.read_csv('instacart_2017_05_01/'+each_file) for each_file in files]
```

# Initial Analysis

```python
products_df=dfs[0]
orders_df=dfs[1]
order_products_train_df=dfs[2]
departments_df=dfs[3]
aisles_df=dfs[4]
order_products_prior_df=dfs[5]
```

```python
products_df.tail(10)
```

```python
print((products_df['product_id'].value_counts()>1).sum()) # check for duplicates in product_id
print((products_df['product_name'].value_counts()>1).sum()) # check for duplicates in product_name

cat_cols=['aisle_id', 'department_id']
fig, ax=plt.subplots(-(-len(cat_cols)//subplot_cols), subplot_cols, figsize=(chart_h*subplot_cols*aspect_ratio, chart_h*(-(-len(cat_cols)//subplot_cols))))
for index, each_col in enumerate(cat_cols): 
    if len(products_df[each_col].unique())<=max_bars: 
        products_df[each_col].value_counts().sort_values(ascending=False).plot(kind='bar', ax=ax[index%subplot_cols])
    else: 
        products_df[each_col].hist(bins=max_bars, ax=ax[index%subplot_cols])
    ax[index%subplot_cols].set_title(f'fig_{index+1}: {each_col}')
```

This table stores the atomic level data of each unique_product

**Observations:**
* There are 49688 unique products
* Every product name is unique
* fig_1: some asiles have 1000+ products
* fig_2: some departments have 6000+ products

```python
orders_df[(orders_df['eval_set'].isin(['train', 'test']))]['order_number'].hist(figsize=(chart_h*aspect_ratio, chart_h))
```

**Observation:**
* There is at least 3 records of historical orders on the train and test set
* For the model training and testing - there is at least 3 days of previous purchases to predict the next

**Observation:**
* There is an increase of customers that placed their first order from nearly a year ago
* There is a peak of users that ordered 90 days ago and to a lesser extend 120 days ago. This would suggests that after a 90 day trial period, people have a tendency to drop off from the service
* There is an increase in users that are greater than 340 days ago

```python
orders_df['cumsum']=orders_df[['order_id', 'user_id', 'days_since_prior_order']].groupby('user_id').cumsum()['days_since_prior_order']
orders_df['cumsum']=orders_df['cumsum'].fillna(0)
orders_df[['user_id', 'cumsum']].groupby('user_id').max().hist(bins=30, figsize=(chart_h*aspect_ratio, chart_h))
```

```python
orders_df[['order_id', 'order_dow', 'order_hour_of_day']].pivot_table(values='order_id', index='order_dow', columns='order_hour_of_day', aggfunc='count')
```

```python
plt.figure(figsize=(chart_h*aspect_ratio, chart_h))
sns.heatmap(orders_df[['order_id', 'order_dow', 'order_hour_of_day']].pivot_table(values='order_id', index='order_dow', columns='order_hour_of_day', aggfunc='count').sort_index(ascending=False))
```

**Observation**
* Saturday evening and sunday mornings are significantly most popular time to shop

```python
import numpy as np

# test_df=orders_df.head(50)
dow=np.sort(orders_df['order_dow'].unique())
fig, ax=plt.subplots(-(-len(dow)//subplot_cols), subplot_cols, figsize=(chart_h*subplot_cols*aspect_ratio, chart_h*(-(-len(dow)//subplot_cols))))
for index, each_dow in enumerate(dow): 
    orders_df[orders_df['order_dow']==each_dow][['user_id', 'order_hour_of_day']].groupby('user_id').std().hist(bins=30, ax=ax[index//subplot_cols][index%subplot_cols])
    ax[index//subplot_cols][index%subplot_cols].set_title(f'fig_{index+1}: {each_dow}')
orders_df[['user_id', 'order_hour_of_day']].groupby('user_id').std().hist(bins=30, ax=ax[-1][-1])
    
```

**Observation**
* Most people shop within the same three hour range

```python
ax=orders_df[['order_dow', 'order_hour_of_day']].boxplot(by='order_dow', figsize=(chart_h*aspect_ratio, chart_h), showfliers=False)
ax.set_title('Frequent Order Hour of Day')
```

**Observation**
* The range of values for hours of the day orders are placed is similar across the days of the week

```python
orders_df[['user_id', 'days_since_prior_order']].groupby('user_id').mean().hist(bins=30)
plt.axvline(orders_df[['user_id', 'days_since_prior_order']].groupby('user_id').mean().mean().iloc[0], color='red')
```

**Observation**
* On average customers reorder every 15 days

```python
ax=orders_df[['order_number', 'days_since_prior_order']].boxplot(by='order_number', figsize=(chart_h*aspect_ratio, chart_h), showfliers=False)
ax=orders_df[['order_number', 'days_since_prior_order']].groupby('order_number').mean().plot(ax=ax)
```

**Observation**
* Overtime customers place orders more frequently (less time between orders). This is however, potentially skewed by a number of users who orders frequently not necessarily based on needs. 

```python
orders_df['cumsum_bin']=pd.cut(orders_df['cumsum'], bins=100)
ax=orders_df[['cumsum_bin', 'days_since_prior_order']].boxplot(by='cumsum_bin', figsize=(chart_h*aspect_ratio, chart_h), showfliers=False)
ax=orders_df[['cumsum_bin', 'days_since_prior_order']].groupby('cumsum_bin').mean().plot(ax=ax)
```

**Observation**
* This is a better depiction of order frequenty - it looks like on average customers do order on a weeky basis
* As users get used to the service, they normalize to an average of around 9 days between orders
